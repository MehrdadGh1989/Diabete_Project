```python
# to check also Whether there is null_value in the DataSet or not and represent in a table
isnull_number = []
for i in diabete_df.columns:
    x = diabete_df[i].isnull().sum()
    isnull_number.append(x)
isnull_number_S=pd.Series(isnull_number,index=diabete_df.columns)
isnull_number_S
```

```python
#To Check How many Values are there in each column
unique_number = []
for i in diabete_df.columns:
    x = diabete_df[i].value_counts().count()
    unique_number.append(x)
unique_number_S=pd.Series(unique_number,index=diabete_df.columns)
unique_number_S
```

```python
#Create a DataFrame of these results (isnull_number_S/unique_number_S)
data={"Total Missing Values":isnull_number_S,"Total_Unique_Values":unique_number_S}
unique_notnull_df=pd.DataFrame(data)
unique_notnull_df
```

```python
#Cleaning_data
#There were no Null_Values in the DataSet
#To understand How Many Rows are duplicated
diabete_df.duplicated(subset=['Diabetes_012','HighBP','HighChol','CholCheck','BMI','Smoker','Stroke',
                              'HeartDiseaseorAttack','PhysActivity','Fruits','Veggies','HvyAlcoholConsump','AnyHealthcare','NoDocbcCost','GenHlth','MentHlth',
                              'PhysHlth','DiffWalk','Sex','Age','Education','Income']).sum()
```

```python
#clean duplicated rows and resetting the index
diabete_df_clean=diabete_df.drop_duplicates()
diabete_df_clean.reset_index(drop=True,inplace=True)
diabete_df_clean
```
